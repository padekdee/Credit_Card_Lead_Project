```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('always')
warnings.filterwarnings('ignore')
```

```python
train = pd.read_csv(r'train_s3TEQDk.csv')
test = pd.read_csv(r'test_mSzZ8RL.csv')
```

__All numerical features and their distribution__

```python
train.hist(figsize = (15, 15))
```

__Checking Skewness__

```python
train.plot(kind='density', subplots=True, \
             layout=(4,4), sharex=False,figsize=(20,15))
plt.show()
```

_Just another way_

```python
for i, column in enumerate(train.select_dtypes(include = [np.int64, np.float64]).columns, 1):
    plt.figure(figsize = (15, 15))
    plt.subplot(3,3,i)
    sns.distplot(train[column])
```

__Checking Outliers__

```python
train.plot(kind = 'box', subplots = True, \
          layout = (3, 5), sharex = False, sharey = False, figsize = (15, 15))
plt.show()
```
